```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.svm import SVC
```

```python
X=data.drop(columns='diagnosis')
```

```python
Y=data['diagnosis']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
num_cols=X.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
num_pipe=Pipeline([
    ("scaler",StandardScaler())
])
```

```python
prep=ColumnTransformer([
    ("num_pipe",num_pipe,num_cols)
],remainder='drop')
```

```python
pipe=Pipeline([
    ("Prep",prep),
    ("model",SVC(kernel='rbf'))
])
```

```python
pipe.fit(X_train,Y_train_scaled)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
print("Accuracy score is",accuracy_score(Y_test_scaled,Y_pred))
print("Classification Report is: \n",classification_report(Y_test_scaled,Y_pred))
print("Confusion Matrix is: \n",confusion_matrix(Y_test_scaled,Y_pred))
```

```python
print("Training Accuracy is",pipe.score(X_train,Y_train_scaled))
print("Testing Accuracy is",pipe.score(X_test,Y_test_scaled))
```

```python
param_grid = {
    "model__kernel": ["linear", "rbf", "poly", "sigmoid"],
    "model__C": [0.1, 1, 10],
    "model__gamma": ["scale", "auto"],   
    "model__degree": [2, 3]              
}
```

```python
pipe1=Pipeline([
    ("Prep",prep),
    ("model",SVC())
])
```

```python
grid=GridSearchCV(
    estimator=pipe1,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_score=grid.best_score_
best_params=grid.best_params_
```

```python
print("Best score is",best_score)
print("Best Parameters are",best_params)
```

```python
Y_pred1=best_model.predict(X_test)
```
