```python
import pandas as pd
```

```python
data=pd.read_csv("advertising.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns='Sales')
```

```python
Y=data['Sales']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing  import StandardScaler
from sklearn.linear_model import Lasso,LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
STD=StandardScaler()
X_train_scaled=STD.fit_transform(X_train)
X_test_scaled=STD.transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
Y_pred=model.predict(X_test_scaled)
```

```python
print("r2 score is",r2_score(Y_test,Y_pred))
print("Mean Squared Error is",mean_squared_error(Y_test,Y_pred))
print("Mean Absolute Error ",mean_absolute_error(Y_test,Y_pred))
```

```python
model1=Lasso(max_iter=10000)
```

```python
model1.fit(X_train_scaled,Y_train)
```

```python
Y_pred1=model1.predict(X_test_scaled)
```

```python
print("Training Accuracy is ",model1.score(X_train_scaled,Y_train))
print("Testing Accuracy is ",model1.score(X_test_scaled,Y_test))
```

```python
print("r2 score is",r2_score(Y_test,Y_pred1))
print("Mean Squared Error ",mean_squared_error(Y_test,Y_pred1))
print("Mean Absolute Error ",mean_absolute_error(Y_test,Y_pred1))
```

```python
grid = GridSearchCV(
    estimator=Lasso(max_iter=10000),
    param_grid={
        "alpha": [0.001, 0.01, 0.1, 1, 10]
    },
    scoring="r2",
    cv=5,
    n_jobs=-1
)
```

```python
grid.fit(X_train_scaled,Y_train)
```

```python
best_model=grid.best_estimator_
best_score=grid.best_score_
best_params=grid.best_params_
```

```python
print("Best Score is",best_score)
print("Best Parameters are",best_params)
```

```python
Y_pred2=best_model.predict(X_test_scaled)
```

```python
print("r2 score is",r2_score(Y_test,Y_pred2))
print("Mean Squared Error is",mean_squared_error(Y_test,Y_pred2))
print("Mean Absolute Error is",mean_absolute_error(Y_test,Y_pred2))
```

```python
print("Training Accuracy is",grid.score(X_train_scaled,Y_train))
print("Testing Accuracy is",grid.score(X_test_scaled,Y_test))
```
