```python
import pandas as pd
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
X=data.drop(columns=['diagnosis'])
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
STD=StandardScaler()
X_train_scaled=STD.fit_transform(X_train)
X_test_scaled=STD.transform(X_test)
```

```python
XGB = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42
)
```

```python
XGB.fit(X_train_scaled,Y_train_scaled)
```

```python
Y_pred=XGB.predict(X_test_scaled)
```

```python
print("Accuracy score is",accuracy_score(Y_test_scaled,Y_pred))
print("Classification Report is: \n",classification_report(Y_test_scaled,Y_pred))
print("Confusion Matrix is: \n",confusion_matrix(Y_test_scaled,Y_pred))
```

```python
print("Training Accuracy is",XGB.score(X_train_scaled,Y_train_scaled))
print("Testing Accuracy is",XGB.score(X_test_scaled,Y_test_scaled))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}
```

```python
grid = GridSearchCV(
    estimator=XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        random_state=42
    ),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train_scaled,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_params=grid.best_params_
best_score=grid.best_score_
```

```python
print("Best score is",best_score)
print("Best Parameters are",best_params)
```

```python
Y_pred1=best_model.predict(X_test_scaled)
```

```python
print("Accuracy score is",accuracy_score(Y_test_scaled,Y_pred1))
print("Classification Report is: \n",classification_report(Y_test_scaled,Y_pred1))
print("Confusion Matrix is: \n",confusion_matrix(Y_test_scaled,Y_pred1))
```

```python
print("Training Accuracy is",grid.score(X_train_scaled,Y_train_scaled))
print("Testing Accuracy is",grid.score(X_test_scaled,Y_test_scaled))
```
