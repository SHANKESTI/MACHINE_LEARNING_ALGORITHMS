```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.sample(10)
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.isna().sum()
```

```python
X=data.drop(columns=['diagnosis'])
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
STD=StandardScaler()
X_train_scaled=STD.fit_transform(X_train)
X_test_scaled=STD.transform(X_test)
```

## BAGGING 

```python
Bagging=BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=100,
    max_samples=1.0,      # 100% of training data
    max_features=1.0,     # 100% of features
    bootstrap=True,       # WITH replacement (Bagging)
    bootstrap_features=False,
    random_state=42
)
```

```python
Bagging.fit(X_train_scaled,Y_train_scaled)
```

```python
Y_pred=Bagging.predict(X_test_scaled)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred)
matrix=confusion_matrix(Y_test_scaled,Y_pred)
report=classification_report(Y_test_scaled,Y_pred)
```

```python
print("Accuracy score is",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```

```python
Bagging1=BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'max_samples': [0.5, 0.7, 0.9, 1.0],
    'max_features': [0.5, 0.7, 0.9, 1.0],
    'bootstrap': [True, False],
    'bootstrap_features': [True, False]
}
```

```python
grid=GridSearchCV(
    estimator=Bagging1,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)
```

```python
grid.fit(X_train_scaled,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
```

```python
Y_pred1=best_model.predict(X_test_scaled)
```

```python
score1=accuracy_score(Y_test_scaled,Y_pred1)
matrix1=confusion_matrix(Y_test_scaled,Y_pred1)
report1=classification_report(Y_test_scaled,Y_pred1)
```

```python
print("Accuracy score is",score1)
print("Confusion Matrix is \n",matrix1)
print("Classification Report is \n",report1)
```

## PASTING

## Random Subspaces

```python
random_subspaces = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=100,
    max_samples=1.0,
    max_features=0.7,     # Only 70% features
    bootstrap=False,
    bootstrap_features=True,  # Bootstrap features
    random_state=42
)
```

```python
Y_predR=random_subspaces.predict(X_test_scaled)
```

```python
print("Accuracy Score is",scorer)
print("Confusion Matrix is \n",matrixr)
print("Classification Report is \n",reportr)
```

## Random Patches

```python
# Both samples AND features are randomly selected
random_patches = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=100,
    max_samples=0.8,      # 80% samples
    max_features=0.7,     # 70% features
    bootstrap=True,       # Bootstrap samples
    bootstrap_features=True,  # Bootstrap features
    random_state=42
)
```

```python
random_patches.fit(X_train_scaled,Y_train_scaled)
```

```python
Y_predRP=random_patches.predict(X_test_scaled)
```

```python
print("Accuracy Score is",scorerp)
print("Confusion Matrix is \n",matrixrp)
print("Classification Report is \n",reportrp)
```
