```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.sample(5)
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.sample(5)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns=['diagnosis'])
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.sample(5)
```

```python
LB=LabelEncoder()
```

```python
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
STD=StandardScaler()
X_train_scaled=STD.fit_transform(X_train)
X_test_scaled=STD.transform(X_test)
```

```python
ADA=AdaBoostClassifier(
    estimator=DecisionTreeClassifier(
        max_depth=1,
        random_state=42
    ),
    n_estimators=100,
    learning_rate=1.0,
    algorithm='SAMME',
    random_state=42
)
```

```python
ADA.fit(X_train_scaled,Y_train_scaled)
```

```python
Y_pred=ADA.predict(X_test_scaled)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred)
matrix=confusion_matrix(Y_test_scaled,Y_pred)
report=classification_report(Y_test_scaled,Y_pred)
```

```python
print("Accuracy score is",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.5, 1.0],
    'estimator__max_depth': [1, 2]
}
```

```python
grid=GridSearchCV(
    estimator=AdaBoostClassifier(
    estimator=DecisionTreeClassifier(
        max_depth=1,
        random_state=42
    )),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train_scaled,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_score=grid.best_score_
best_param=grid.best_params_
```

```python
print("Best score is",best_score)
print("Best parameters are ",best_param)
```

```python
Y_pred1=best_model.predict(X_test_scaled)
```

```python
score1=accuracy_score(Y_test_scaled,Y_pred1)
matrix1=confusion_matrix(Y_test_scaled,Y_pred1)
report1=classification_report(Y_test_scaled,Y_pred1)
```

```python
print("Accuracy score is",score1)
print("Confusion Matrix is \n",matrix1)
print("Classification Report is \n",report1)
```
