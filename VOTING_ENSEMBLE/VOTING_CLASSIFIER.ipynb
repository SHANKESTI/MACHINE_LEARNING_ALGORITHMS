```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.sample(5)
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns='diagnosis')
Y=data['diagnosis']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression 
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
num_cols=X.select_dtypes(include=['int64','float64']).columns
```

```python
Pipe1=Pipeline([
    ("Prep",ColumnTransformer([("Scaler",StandardScaler(),num_cols)],remainder='passthrough')),
    ("LG",LogisticRegression())
])
```

```python
Pipe2=Pipeline([
    ("Prep",ColumnTransformer([("Scaler",StandardScaler(),num_cols)],remainder='passthrough')),
    ("KNN",KNeighborsClassifier())
])
```

```python
Pipe3=Pipeline([
    ("Prep",ColumnTransformer([("Scaler",StandardScaler(),num_cols)],remainder='passthrough')),
    ("GNB",GaussianNB())
])
```

## SOFT VOTING CLASSIFIER

```python
Voting=VotingClassifier(
    estimators=[
        ("LG",Pipe1),
        ("KNN",Pipe2),
        ("GNB",Pipe3)
    ],
    voting='soft'
)
```

```python
Voting.fit(X_train,Y_train_scaled)
```

```python
Y_pred=Voting.predict(X_test)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred)
matrix=confusion_matrix(Y_test_scaled,Y_pred)
report=classification_report(Y_test_scaled,Y_pred)
```

```python
print("Accuracy score is",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```

```python
param_grid = {
    'LG__LG__C': [0.01, 0.1, 1, 10],

    'KNN__KNN__n_neighbors': [3,5,7,11],
    'KNN__KNN__weights': ['uniform', 'distance'],

    'GNB__GNB__var_smoothing': [1e-9, 1e-8, 1e-7]
}
```

```python
grid=GridSearchCV(
    estimator=Voting,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
grid.fit(X_train,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_score=grid.best_score_
best_params=grid.best_params_
```

```python
print("Best Parametes are \t",best_params)
print("Best score is \t",best_score)
```

```python
Y_pred1=best_model.predict(X_test)
```

```python
score1=accuracy_score(Y_test_scaled,Y_pred1)
matrix1=confusion_matrix(Y_test_scaled,Y_pred1)
report1=classification_report(Y_test_scaled,Y_pred1)
```

```python
print("Accuracy score is",score1)
print("Confusion matrix is\n",matrix1)
print("Classification report is\n",report1)
```
