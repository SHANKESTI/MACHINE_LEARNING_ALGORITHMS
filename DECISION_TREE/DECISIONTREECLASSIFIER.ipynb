```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.drop(columns=['Unnamed: 32','id'])
```

```python
data.sample(5)
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns='diagnosis')
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
num_cols=X.select_dtypes(include=['int64','float64']).columns
```

```python
trf1=ColumnTransformer([
    ("Scaler",StandardScaler(),num_cols)
    
],remainder='passthrough')
```

```python
trf2=DecisionTreeClassifier()
```

```python
pipe=Pipeline([
    ("trf1",trf1),
    ("trf2",trf2)
])
```

```python
pipe.fit(X_train,Y_train_scaled)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred)
matrix=confusion_matrix(Y_test_scaled,Y_pred)
report=classification_report(Y_test_scaled,Y_pred)
```

```python
print("Accuracy score is",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```

```python
param_grid = {
    'trf2__max_depth': [3,5,7,10,15,20,None],
    'trf2__min_samples_split': [2,5,10,20,50],
    'trf2__min_samples_leaf': [1,2,5,10,20],
    'trf2__max_features': [None, 'sqrt', 'log2'],
    'trf2__criterion': ['gini', 'entropy']
}
```

```python
grid=GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)
```

```python
grid.fit(X_train,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_params=grid.best_params_
best_score=grid.best_score_
```

```python
print(best_params)
print(best_score)
```

```python
Y_pred1=best_model.predict(X_test)
```

```python
score1=accuracy_score(Y_test_scaled,Y_pred1)
matrix1=confusion_matrix(Y_test_scaled,Y_pred1)
report1=classification_report(Y_test_scaled,Y_pred1)
```

```python
print("Accuracy score is",score1)
print("Confusion Matrix is \n",matrix1)
print("Classification Report is \n",report1)
```
