```python
import pandas as pd
```

```python
data=pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['id','Unnamed: 32'])
```

```python
data.sample(5)
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns=['diagnosis'])
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import StackingClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
STD=StandardScaler()
X_train_scaled=STD.fit_transform(X_train)
X_test_scaled=STD.transform(X_test)
```

```python
base_models = [
    ('lr', LogisticRegression(max_iter=1000)),
    ('knn', KNeighborsClassifier()),
    ('gnb', GaussianNB())
]
```

```python
meta_model=LogisticRegression()
```

```python
stack_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=5,
    n_jobs=-1
)
```

```python
stack_model.fit(X_train_scaled, Y_train_scaled)

```

```python
Y_pred = stack_model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(Y_test_scaled,Y_pred))
print("Confusion Matrix is:\n",confusion_matrix(Y_test_scaled,Y_pred))
print("Classification Report is:\n",classification_report(Y_test_scaled,Y_pred))
```

```python
param_grid = {
    'lr__C': [0.01, 0.1, 1, 10],

    'knn__n_neighbors': [3, 5, 7, 11],
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2],   

    
    'gnb__var_smoothing': [1e-9, 1e-8, 1e-7],

    
    'final_estimator__C': [0.01, 0.1, 1, 10]
}
```

```python
grid=GridSearchCV(
    estimator=stack_model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train_scaled,Y_train_scaled)
```

```python
best_model=grid.best_estimator_
best_score=grid.best_score_
best_params=grid.best_params_
```

```python
print("Best Score is",best_score)
print("Best Parameters are",best_params)
```

```python
Y_pred1=best_model.predict(X_test_scaled)
```

```python
print("Accuracy Score is",accuracy_score(Y_test_scaled,Y_pred1))
print("Classification Report is \n",classification_report(Y_test_scaled,Y_pred1))
print("Confusion Matrix is \n",confusion_matrix(Y_test_scaled,Y_pred1))
```
