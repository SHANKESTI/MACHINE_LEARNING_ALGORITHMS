```python
import pandas as pd
```

```python
data=pd.read_csv("breast-cancer.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns=['id','diagnosis'])
```

```python
X.head()
```

```python
Y=data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
num_cols=X.select_dtypes(include=['int64','float64']).columns
```

```python
LB=LabelEncoder()
Y_train_scaled=LB.fit_transform(Y_train)
Y_test_scaled=LB.transform(Y_test)
```

```python
trf1=ColumnTransformer([
    ("Scaler",StandardScaler(),num_cols)
],remainder='passthrough')
```

```python
trf2=KNeighborsClassifier()
```

```python
pipe=Pipeline([
    ("trf1",trf1),
    ("trf2",trf2)
])
```

```python
pipe.fit(X_train,Y_train_scaled)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred)
matrix=confusion_matrix(Y_test_scaled,Y_pred)
report=classification_report(Y_test_scaled,Y_pred)
```

```python
print("Accuracy Score is ",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```

```python
param_grid = {
    'trf2__n_neighbors': [3,5,7,9,11,15,21,25],
    'trf2__weights': ['uniform', 'distance'],
    'trf2__metric': ['euclidean', 'manhattan', 'minkowski'],
    'trf2__p': [1, 2]
}
```

```python
grid=GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
grid.fit(X_train,Y_train_scaled)
```

```python
best_model = grid.best_estimator_
```

```python
best_params = grid.best_params_
print(best_params)
```

```python
best_score = grid.best_score_
print(best_score)
```

```python
Y_pred1=best_model.predict(X_test)
```

```python
score=accuracy_score(Y_test_scaled,Y_pred1)
matrix=confusion_matrix(Y_test_scaled,Y_pred1)
report=classification_report(Y_test_scaled,Y_pred1)
```

```python
print("Accuracy Score is",score)
print("Confusion Matrix is \n",matrix)
print("Classification Report is \n",report)
```
